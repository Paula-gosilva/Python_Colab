```python
df.columns = ['ano', 'senioridade', 'contrato', 'cargo', 'salario', 'moeda', 'usd', 'residencia', 'remoto', 'empresa', 'tamanho_empresa']
print(df.columns)
```

```python
senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
display(df['senioridade'].value_counts())
```

```python
contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Per√≠odo',
    'FL': 'Freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
display(df['contrato'].value_counts())
```

```python
tamanho_empresa = {
    'M': 'M√©dio',
    'L': 'Grande',
    'S': 'Pequeno'
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
display(df['tamanho_empresa'].value_counts())
```

```python
remoto = {
    0: 'Presencial',
    50: 'H√≠brido',
    100: 'Remoto'
}
df['remoto'] = df['remoto'].replace(remoto)
display(df['remoto'].value_counts())
```

```python
import numpy as np

#Cria√ß√£o de um dataFrame de teste parausar como exemplo
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
 })

#calcula a m√©dia e substitui os nulos pela m√©dia e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#calcula mediana e substitui os nulos pela mediana
df_salarios['salario_medina'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia' : ['Segunda', 'Ter√ßa', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura' : [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()

df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dia' : ['Segunda', 'Ter√ßa', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura' : [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade' : ['S√£o Paulo', 'np.nan', 'Curitiba', 'np.nan', 'Bel√©m'],
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].replace('np.nan', 'N√£o informado')
df_cidades
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data = df_limpo, x = 'senioridade', y = 'usd')
plt.title('Sal√°rio m√©dio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Sal√°rio m√©dio anual (USD)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data = df_limpo, x = 'senioridade', y = 'usd', order=ordem)
plt.title('Sal√°rio m√©dio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Sal√°rio m√©dio anual (USD)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df_limpo['usd'], bins = 50, kde=True)
plt.title('Distrubui√ß√£o dos sal√°rios anuais')
plt.xlabel('Sal√°rio em USD')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_limpo['usd'])
plt.title('Boxplot Sal√°rio')
plt.xlabel('Sal√°rio em USD')
plt.show()
```

```python
ordem_senioridade = df_limpo['senioridade'].unique()

plt.figure(figsize=(8,5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade)
plt.title("Boxplot da distribui√ß√£o por senioridade")
plt.xlabel("Sal√°rio em USD")
plt.show()
```

```python
import plotly.express as px
```

```python
salario_por_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reset_index()
salario_por_senioridade = salario_por_senioridade.sort_values('usd', ascending=False)

fig = px.bar(salario_por_senioridade,
             x='senioridade',
             y='usd',
             title='M√©dia Salarial por Senioridade',
             labels={'senioridade': 'Senioridade', 'usd': 'M√©dia Salarial Anual (USD)'})
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Propor√ß√£o dos tipos de trabalho')

fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Propor√ß√£o dos tipos de trabalho',
             hole=0.5)

fig.show()
```
